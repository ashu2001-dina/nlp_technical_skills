```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.util import ngrams
lemmatizer = WordNetLemmatizer()
from collections import Counter
```

```python
df=pd.read_csv('/Users/ahswati/Desktop/Files/Raw_Skills_Dataset.csv')
df.head(10)
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
import regex as re
df['RAW DATA']=df['RAW DATA'].str.lower()
df['RAW DATA']=df['RAW DATA'].apply(lambda x:re.sub(r'[^\w\d\s]+','',x))

df.head()
```

```python
df['RAW DATA']=df['RAW DATA'].apply(word_tokenize)
df.head()
```

```python
stop_words = set(stopwords.words('english'))
```

```python
df['RAW DATA']=df['RAW DATA'].apply(lambda x:[word for word in x if word not in stop_words])
df['RAW DATA']=[' '.join(map(str,l)) for l in df['RAW DATA']]
df.head(10)
```

```python
df['RAW DATA']=[lemmatizer.lemmatize(word) for word in df['RAW DATA'] if not word in stop_words]
```

```python
df.head(10)
```

```python
li=df['RAW DATA'].tolist()
c=Counter(li).most_common()
c
```

```python
s=''
for i in df['RAW DATA']:
    s+=" "+i
```

```python
tokens = [token for token in s.split(" ") if token != ""]
output = list(ngrams(tokens, 2))
```

```python
SKILLS = [
    'python',
    'word',
    'excel',
    'English',
    'azure',
    'mysql',
    'kubernetes'
]
 
def extract_skills(input_text):
    
    found_skills = set()
    for token in li:
        if token in SKILLS:
            found_skills.add(token)
    for ngram in output:
        if ngram in SKILLS:
            found_skills.add(ngram)
            
    return found_skills
 
skills=extract_skills(li)
```
